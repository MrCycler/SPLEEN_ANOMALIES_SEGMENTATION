```
import glob
import re

#obtiene los paths para entrenamiento
paths_tr = glob.glob("Task09_Spleen/imagesTr/*.nii.gz")
numbers_tr = list(map(lambda x: str(int(re.search(r'\d+', x.split('spleen_')[1]).group(0))), paths_tr))

#obtiene los paths para validacion
paths_ts = glob.glob("Task09_Spleen/imagesTs/*.nii.gz")
numbers_ts = list(map(lambda x: str(int(re.search(r'\d+', x.split('spleen_')[1]).group(0))), paths_ts))
```

```
# https://github.com/alexlaurence/NIfTI-Image-Converter/blob/master/python/nii2png.py
import convert_v2 as cvt

#extrae las imagenes de training y las guarda en la carpeta
for i in range(len(paths_tr)):
  cvt.convert(paths_tr[i],'images_tr/'+numbers_tr[i])

#extrae las imagenes de test y las guarda en la carpeta
for j in range(len(paths_ts)):
  cvt.convert(paths_ts[j],'images_ts/'+numbers_ts[j]) 
```
